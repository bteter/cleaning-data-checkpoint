## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
df.isna() #checks for null values, creates dataframe where all values are replaced with a boolean value where null is True, otherwise False.
#at a quick glance, there are no obvious values.
```

```python
df.isna().sum() #gives sum of all null values per column.
#dataset only has one null value
```

```python
#to get rid of that null value, I want to replace the value with the average value from this column.

x = df["us_viewers_in_millions"].mean() # x = the average viwership number

df.fillna(x) # fillna replaces any NaN values in the DataFrame with the average viewership number
```

```python
df.isna().sum() #double check to see if null value has been replace.

#us_viewers_in_millions has 0 null values.
```
